#### Feature Engineering

```python
import pandas as pd
import numpy as np
import category_encoders as ce
```

```python
df = pd.read_csv('post_eda.csv')
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.head()
```

```python
df['Vehicle_condition'] = df['Vehicle_condition'].astype('object')

num_cols = [col for col in df.columns if df[col].dtype != 'O']
cat_cols = [col for col in df.columns if df[col].dtype == 'O']

print('num_cols : {}\n'.format(num_cols))
print('cat_cols : {}\n'.format(cat_cols))
```

```python
df[cat_cols].head()
```

```python
df[['Order_Date_Day', 'Order_Date_Month', 'Order_Date_Year']] = pd.DataFrame(df['Order_Date'].str.split('-', expand=True))

df.drop(['Order_Date', 'Order_Date_Year'], axis=1, inplace=True)

df.head()
```

```python
df[['Time_Orderd_Hr', 'Time_Orderd_Min']] = pd.DataFrame(df['Time_Orderd'].str.split(':', expand=True))

df[['Time_Order_picked_Hr', 'Time_Order_picked_Min', 'Time_Order_picked_Sec']] = pd.DataFrame(df['Time_Order_picked'].str.split(':', expand=True))

df.drop(['Time_Order_picked', 'Time_Orderd', 'Time_Order_picked_Sec'], axis=1, inplace=True)

df.head()
```

```python
num_cols = [col for col in df.columns if df[col].dtype != 'O']
cat_cols = [col for col in df.columns if df[col].dtype == 'O']
```

```python
df[cat_cols].head()
```

```python
def do_binary_encoding(features:list(), df:pd.DataFrame)->(pd.DataFrame, list()):
    
    binary_encoder = ce.BinaryEncoder()

    new_added_features = []

    for feature in features:

        new_feature_list = list(pd.DataFrame(binary_encoder.fit_transform(df[feature])).columns)
        df[new_feature_list] = pd.DataFrame(binary_encoder.fit_transform(df[feature]))

        new_added_features += new_feature_list
                                            
        df.drop(feature, axis=1, inplace=True)

    return (df.copy(deep=True), new_added_features)
```

```python
encoded_df, newly_added_features = do_binary_encoding(cat_cols, df=df)
```

```python
encoded_df[newly_added_features].head()
```

### Model Trainning and Prediction

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.metrics import r2_score
```

```python
X = df.drop(columns=['Time_taken (min)'], axis=1)

y = df['Time_taken (min)']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=15)
```

```python
standard_scalar = StandardScaler()

X_train = standard_scalar.fit_transform(X_train)
X_test = standard_scalar.transform(X_test)
```

```python
linear_regression = LinearRegression()
ridge = Ridge()
elastic_net = ElasticNet()
lasso = Lasso()
```

```python
linear_regression.fit(X_train, y_train)
ridge.fit(X_train, y_train)
lasso.fit(X_train, y_train)
elastic_net.fit(X_train, y_train)
```

```python
y_pred = linear_regression.predict(X_test)

r2_score(y_true=y_test, y_pred=y_pred) * 100
```

```python
y_pred = ridge.predict(X_test)

r2_score(y_true=y_test, y_pred=y_pred) * 100
```

```python
y_pred = elastic_net.predict(X_test)

r2_score(y_true=y_test, y_pred=y_pred) * 100
```

```python
y_pred = lasso.predict(X_test)

r2_score(y_true=y_test, y_pred=y_pred) * 100
```

```python
parameters = {
    'fit_intercept' : [True, False],
    'copy_X'        : [True, False],
    'positive'      : [True, False]
}
```

```python
grid_GBR = GridSearchCV(estimator=linear_regression, param_grid=parameters, cv=2, n_jobs=-1)

grid_GBR.fit(X_train, y_train)
```

```python
print('Results of Grid Search CV : ')
print("\n The best estimator across ALL searched params:\n {}".format(grid_GBR.best_estimator_))
print("\n The best score across ALL searched params:\n {}".format(grid_GBR.best_score_))
print("\n The best parameters across ALL searched params:\n {}".format(grid_GBR.best_params_))
```
